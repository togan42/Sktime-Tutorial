Prerequisites:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

import sktime
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.compose import make_reduction
from sktime.forecasting.trend import TrendForecaster,PolynomialTrendForecaster
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.transformations.series.detrend import Detrender , Deseasonalizer
from sktime.forecasting.naive import NaiveForecaster
from sktime.forecasting.theta import ThetaForecaster
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.forecasting.base import ForecastingHorizon

# plotting modules:
from sktime.utils.plotting import plot_series,plot_lags, plot_correlations
from sktime.utils.plotting import plot_series

```

# 1. From a Csv file to an sktime compatible data format:

Dataset contains monthly production of beer in Australia.

Resource: https://www.kaggle.com/datasets/sergiomora823/monthly-beer-production


```python
# Reading the csv file
df = pd.read_csv("monthly_beer.csv")
df.head()
```

Time series data starts from 1956-01 and has monthly frequency

```python
# Creating variable y and giving it a proper index according to the data
y = df["Monthly beer production"]
y.index = pd.date_range("1956-01", freq="M", periods=len(df))
# Important!  The index should be in pandas PeriodIndex format to be used with sktime
y.index = y.index.to_period()
y[:3]
```

```python
# A first look at data
plot_series(y)
```

# 2. Time series Analysis:

Analysis of time series components will be made. 
Our analysis will include trend, seasonality and cyclic behaviour of time series data. Autocorrelation and lag plots are helpful tools to understand these components.

### 2.1 Autocorrelation plots:
Autocorrelation plots show the correlation between the elements of a time series. The time period between an element and the previous one is called lag. The plot consists of these lags in x-axis and coefficient of correation in y-axis.

```python
plot_correlations(y)
```

From both autocorrelation plots, it can be seen that the time series has a strong correlation in lag-12. And it repeats in the first plot(lag-24). This can be a proof of an annual seasonality of our monthly data.

```python
# We can see the strong correlation of lag-12 clearly with lag plots:
plot_lags(y , lags=[1,6,12])
```


### 2.2 Seasonal decomposition with statsmodels:
To investigate every component of a time series, the data can be decomposed in pieces.

```python
# The dataframe format that statsmodel needs for seasonal decompose module:
df2 = pd.read_csv("monthly_beer.csv" , parse_dates=True , index_col="Month")
# Composing components by "adding" them. So the model is additive
result = seasonal_decompose(df2["Monthly beer production"] , model="add")
residuals = result.resid
result.plot()
plt.show()
```

#### Stationarity:

Stationarity of time series depends on whether its properties vary over time. If it varies then time series called non-stationary. A time series with a strong trend or seasonality is non-stationary. It means these components will change the value of the time series at different times.

Stationary time series are ideal for modelling. By using different transformations, a non-stationary time series can be made stationary.


Stationarity of a time series can be tested using Dickey-Fuller test. Stationarity limits for Dickey-Fuller test:
* Lower test statistic value then critical values.
* Lower p-value than 0.05.

```python
# Result printing function:
def test_adfuller(ts):
    result = adfuller(ts, autolag='AIC')
    print('Test statistic: ' , result[0])
    print('p-value: '  ,result[1])
    print('Critical Values:' ,result[4])
```

```python
print("Original data result:")
test_adfuller(y)
print("\nAfter decomposition:")
test_adfuller(residuals.dropna())
```

It shows that statsmodels decomposition extracted enough information to make the time series stationary

### 2.3 Seasonal decomposition with sktime:

First, detrending the time series:

```python
# Forecaster and transformers are defined:
forecaster = PolynomialTrendForecaster(degree=5) 
transformer = Detrender(forecaster=forecaster)
detrended = transformer.fit_transform(y)

# To make an in-sample forecasting horizon use minus range(it will be positive range for forecasting):
fh_in = -np.arange(len(y))
# The trend line:
y_pred = forecaster.fit(y).predict(fh=fh_in)
```

```python
plot_series(y,detrended,y_pred , labels=["Original data","Residuals","Fitted line"])
```

Secondly, Deseasonalizing the rest series: 

```python
# model is created
deseasonalizer = Deseasonalizer(model="additive", sp=12)
# Is fitted into detrended time series
resid = deseasonalizer.fit_transform(detrended)
```

```python
plot_series(y,resid,labels=["Original data","Residuals"])
```

```python
print("Original data test result:")
test_adfuller(y)
print("\nAfter detrending test result:")
test_adfuller(detrended)
print("\nAfter detrending and deseasoning test result:")
test_adfuller(resid)
```

Better result after each transformation. 

# 3. Forecasting:

Splitting data into training and testing sets:

```python
y_train , y_test = temporal_train_test_split(y , test_size=48 )
```

## 3.1 Simple forecasting with a built-in sktime forecaster:

```python
# Training part:
# Define a forecaster and fit training data
forecaster = ThetaForecaster(sp=12)
forecaster.fit(y_train)

# Testing part:
# Create a forecasting horizon and predict in that interval
fh = np.arange(1, 49)
y_pred = forecaster.predict(fh)

print("Mean Absolute Percentage Error: %" , 100*mean_absolute_percentage_error(y_pred , y_test))
```

## 3.2 Forecasting with make_reduction method:
To make forecasting task with a sklearn regressor, make_reduction method can be used. 

The sktime data is reduced into a sklearn trainable tabular data.

```python
# Regressors:
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.neighbors import KNeighborsRegressor
#from xgboost import XGBRegressor
```

### 3.2.1 Direct forecasting strategy

In the direct strategy, the model is trained to forecast a fixed length of interval.
So this length information should be given before training.


```python
# Regressor instance and the forecaster are created.
regressor =  LinearRegression()
# The sytrategy is defined in parameters
forecaster = make_reduction( estimator=regressor  , strategy = "direct" , window_length=12 ) # default: window_length=10
fh = ForecastingHorizon(np.arange(1,49))
# !Important: forecasting horizan is given:
forecaster.fit(y=y_train , fh=fh)
```

```python
# Forecasting:
y_pred_dir = forecaster.predict(fh=fh)
print("Mean Absolute Percentage Error: %" , 100*mean_absolute_percentage_error(y_pred_dir , y_test))
```

### 3.2.2 Recursive Forecasting Strategy

In this strategy, the model is trained to only guess the next one data point. 
The guessed data is used to forecast the next one.

```python
regressor =  Ridge()
forecaster = make_reduction(regressor , strategy = "recursive"  , window_length=24 )
fh = ForecastingHorizon(np.arange(1,49))
# no need to define a forecasting horizon before training
forecaster.fit(y=y_train)
```

```python
y_pred_rec = forecaster.predict(fh=fh)
print("Mean Absolute Percentage Error: %" , 100*mean_absolute_percentage_error(y_pred_rec , y_test))
```

Plotting time series result:

```python
plot_series(y_train[200:], y_test, y_pred_dir, labels=["y_train", "y_test", "forecast"])
```
